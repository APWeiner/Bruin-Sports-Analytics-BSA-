```python
class Model():
    def __init__(self, coefficients, intercept):
        self.coefficients = coefficients
        self.intercept = intercept
    def predict(self, X):
        return X.dot(self.coefficients) + self.intercept
```

```python
SalaryPredictor = Model(coefficients=[3.815718562111416,2.32091090376625,-0.034417153618212605,3.895412777294307,-1.1985643975295406  ], intercept=10.043764045550192)
```

```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
current_season_df = pd.read_csv("current stats and salaries.csv")
```

```python
features = ['PTS', 'TOV', 'AST', 'VORP', 'PER']
X = current_season_df[features]
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```

```python
predictions = (SalaryPredictor.predict(X_standardized)*1000000).astype(int)
```

```python
current_season_df['Prediction'] = predictions
```

```python
display_df = current_season_df[['Player','Salary','Prediction']]
```
